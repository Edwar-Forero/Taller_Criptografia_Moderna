# Taller AES: Criptografía Moderna

```python
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad
from Crypto.Util import Counter
```

## Punto 1: Cifrado AES con modo ECB


```python
texto_plano = b"texto secreto"
clave_cifrado = b'A\x83\xbeU\xd7\xa9b\x18\x85AN0\xbft\xc3\xab'
```

### Cifrados sin padding

```python
try:
    cifrar = AES.new(clave_cifrado, AES.MODE_ECB)
    texto_cifrado = cifrar.encrypt(texto_plano)
    print("texto cifrado sin padding:", texto_cifrado)
except ValueError as e:
    print("Error al cifrar:", e)
```

### Cifrados con padding

```python
cifrarp = AES.new(clave_cifrado, AES.MODE_ECB)
padded = pad(texto_plano, 16)
texto_cifradop = cifrarp.encrypt(padded)
print("texto cifrado con padding:", texto_cifradop)
print("Longitud del texto cifrado con padding:", len(texto_cifradop))
```

## Punto 2: Cifrado AES con modo CTR

```python
try:
    cifrar = AES.new(clave_cifrado, AES.MODE_CTR)
    texto_cifrado = cifrar.encrypt(texto_plano)
    print("texto cifrado sin padding:", texto_cifrado)
    print("tamaño del texto cifrado: ", len(texto_cifrado))
except ValueError as e:
    print("Error al cifrar:", e)
```

## Punto 3: Tamaño de clave a 256 bits

```python
clave_cifrado_256 = b'12345678901234567890123456789012'
print("Longitud de la clave de cifrado:", len(clave_cifrado_256))
```

### Modo ECB 

```python
cifrarp256 = AES.new(clave_cifrado_256, AES.MODE_ECB)
padded = pad(texto_plano, 16)
texto_cifradop = cifrarp256.encrypt(padded)
print("texto cifrado con padding:", texto_cifradop)
print("Longitud del texto cifrado con padding:", len(texto_cifradop))
```

### Modo CTR

```python
try:
    cifrar256 = AES.new(clave_cifrado_256, AES.MODE_CTR)
    texto_cifrado = cifrar256.encrypt(texto_plano)
    print("texto cifrado sin padding:", texto_cifrado)
    print("tamaño del texto cifrado: ", len(texto_cifrado))
except ValueError as e:
    print("Error al cifrar:", e)
```

### Descifrar ECB

```python
aes_ecb = AES.new(clave_cifrado, AES.MODE_ECB)
texto_ecb = aes_ecb.decrypt(texto_cifrado_1_ECB)
print("Texto descifrado ECB:", texto_ecb)
```

### Descifrar CTR

```python
ctr = Counter.new(128)
texto_ctr = AES.new(clave_cifrado, AES.MODE_CTR, counter=ctr).decrypt(texto_cifrado_2_CTR)
print("Texto descifrado CTR con Counter:", texto_ctr)
```
